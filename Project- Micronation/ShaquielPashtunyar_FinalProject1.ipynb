```python
#Combining all the former tables
Former = pd.concat([Australia, Austria,Bahamas, Brazil, Canada, CzechRepublic, Germany, Indonesia, Italy, Jamaica, Nigeria, SpratlyIsland, Tanganyika, Tonga, UnitedKingdom, UnitedStates], axis=0)
```

```python
Former["Status"]="Former"
```

```python
Former= Former.drop(columns=["Flag"])
```

```python
Former['YearStart'] = Former['Date'].str[:4]
```

```python
Former['YearEnd'] = Former['Date'].str[5:]
```

```python
Former= Former.replace('','0')
```

```python
Former["YearStart"] = Former["YearStart"].astype(str).astype(int)
```

```python
Former["YearEnd"] =Former["YearEnd"].astype(str).astype(int)
```

```python
Former['Length(In Years)']= Former["YearEnd"] - Former["YearStart"]
```

```python
Former= Former.replace(-1993,0)
Former= Former.replace(-1980,0)
Former= Former.replace(-2020,0)
```

```python
#Fixing Current Data Set
```

```python
Current["Status"]= "Current"
```

```python
Current= Current.drop(columns=["Flag"])
```

```python
Current['YearStart'] = Current['Date'].str[:4]
```

```python
Current= Current.replace('Ca. ','1965')
```

```python
Current['YearEnd'] = 2023
```

```python
Current["YearStart"] = Current["YearStart"].astype(str).astype(int)
```

```python
Current["YearEnd"] =Current["YearEnd"].astype(str).astype(int)
```

```python
Current['Length(In Years)']= (Current["YearEnd"] - Current["YearStart"]).astype(str).astype(int)
```
